## create simple_medians DataFrame

```python
def get_n_days_median(new_df, df,  n):
    """add new column to the new_df where values are n-median of the df"""
    col_number = len(df.columns)
    columns = list(df.columns[col_number-n:col_number])
    
    resultFrame = pd.DataFrame(df.loc[:, columns])
    resultFrame['Median'] = resultFrame.loc[:, :].median(axis=1)
    
    pattern_label = 'Median {} day/days'.format(n)
    new_df[pattern_label] = resultFrame.loc[:, 'Median'].values
```

```python
l = len(df_train.columns.values) + 1
rang = range(1, l)

# create DataFrame
simple_medians = pd.DataFrame()
simple_medians['Page'] = df_train.loc[:,'Page'].values

#add new columns to the DataFrame
for i in rang:
    get_n_days_median(simple_medians, df_train, i)

# set 'Page' column as an index
simple_medians.set_index('Page')
simple_medians.head()
```

## add medians to the df_test

```python
from copy import copy

# create list of the median columns
labels = ['Median {} day/days'.format(i) for i in rang]

# create lists with the const and variable columns name
columns_test = df_test.columns.values[1:]

new_labels = copy(labels)
new_labels.append('Page')

for label in labels:
    df_test[label] = simple_medians.loc[:, label].values

# melt df_test
df_test = pd.melt(df_test, id_vars=new_labels, value_vars=columns_test, var_name='Date', value_name='Visits')
```

## change data format in the df_test

```python
df_test = df_test.set_index('Page')
df_test['Date'] = pd.DatetimeIndex(df_test['Date'])
df_test.head()
```

## Count SMAPE for the simple_medians

```python
def pandas_smape(df, compare_with, new_df, new_label):
    """function to count the accuracy of the prediction"""
    new_df['Visits'].fillna(0, inplace=True)
    df[compare_with].fillna(0, inplace=True)
    new_df[new_label] = 200 * np.abs(new_df["Visits"] - df[compare_with]) / (new_df["Visits"] + df[compare_with])
    new_df[new_label].fillna(0, inplace=True)
    return np.mean(new_df[new_label])
```

```python
smape_df = df_test.loc[:, ['Date', 'Visits']]

smapes = dict()
for index, label in enumerate(labels):
    smapes[rang[index]] = pandas_smape(df_test, label, smape_df, label)
```

```python
minimum= sorted(list(smapes.items()), key=lambda x: x[1])[0]
'Max accuracy by: {}: {}'.format(minimum[0], minimum[1])
```
